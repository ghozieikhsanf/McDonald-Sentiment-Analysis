# II. Import Libraries

```python
import pickle
import pandas as pd

from sklearn.base import BaseEstimator, TransformerMixin
from sentence_transformers import SentenceTransformer
```

# III. Data Loading

```python
with open('embedding_classification.pkl', 'rb') as model_file:
    model=pickle.load(model_file)
    
embedder = SentenceTransformer('all-MiniLM-L6-v2')
```

# IV. Add New Data

```python
new=['food is bad','good but expensive','very worth the price','store is clean, hospitality is really good. but the price will starved me till next week', 'store is really dirty, waiters are impolite',"food is really good but I got kicked out from the store because they don't like my dog"]
new
```

# V. Data Prediction

Embedding

```python
new_embed = embedder.encode(new, show_progress_bar=True)
```

Prediction

```python
y_pred_inf = model.predict(new_embed)
y_pred_inf
```
